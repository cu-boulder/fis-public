```python
from SPARQLWrapper import SPARQLWrapper, JSON
from collections import namedtuple
import json
import requests

endpoint_url = "https://query.wikidata.org/sparql"
ES_person_endpoint_url = "https://experts.colorado.edu/es/fis/person/_search?size=5000"
ES_pub_endpoint_url = "https://experts.colorado.edu/es/fis/person/_search"
headers = {'Content-type': 'application/json'}
orcid = ""
```

```python
def WdataSearch(endpoint_url, query):
    sparql = SPARQLWrapper(endpoint_url)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()
```

```python
def ESsearch(uri):
    """Simple Elasticsearch Query"""
    query = json.dumps({
        "query": {
           "exists": {
                "field": "orcid"
            }

        }
    })
    response = requests.get(uri, headers=headers, data=query)
    results = json.loads(response.text)
    return results

```

```python
print ("Only displaying records with an ORCID")
ESresults = ESsearch(ES_person_endpoint_url)
#print (ESresults)
for eresult in ESresults['hits']['hits']:
        #print (eresult)
        orcid=eresult["_source"]["orcid"]
        query="""SELECT ?item WHERE { ?item wdt:P496 \"""" + orcid + """\" } """
        Wresults = WdataSearch(endpoint_url, query)
        wentity = ""
        for result in Wresults["results"]["bindings"]:
            wentity = result["item"]["value"]
        print (eresult["_source"]["orcid"], eresult["_source"]["name"],  eresult["_source"]["uri"], wentity)
```
